## DATA EXPLORING

### Import library

```python
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import os
```

```python
image_path = "../dataset/train"
```

```python
labels = os.listdir(image_path)
print("Total class:", len(labels))
```

```python
n_diseases = 0
for label in labels:
    if label.split('___')[1] != 'healthy':
        n_diseases += 1

print("Number of diseases:", n_diseases)
```

```python
nums = {}
for disease in labels:
    nums[disease] = len(os.listdir(image_path + '/' + disease))

img_per_class = pd.DataFrame(nums.values(), index=nums.keys(), columns=["no. of images"])
img_per_class
```

```python
index = [i for i in range(10)]
plt.figure(figsize = (9, 4))
plt.bar(index, [n for n in nums.values()], width = 0.3)
plt.xlabel('Diseases', fontsize = 10)
plt.ylabel('Number of images', fontsize = 10)
plt.xticks(index, labels, fontsize = 8, rotation = 90)
plt.title('Images per each class');
```

```python
plt.figure(figsize=(17, 8))
for i, disease in enumerate(labels):
    plt.subplot(2, 5, i + 1)
    list_image = os.listdir(image_path + '/' + disease)
    image = cv2.imread(os.path.join(os.path.join(image_path, disease), list_image[21]))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(disease, fontsize=10)
    plt.axis("off")
```

```python
n_train = 0
for value in nums.values():
    n_train += value
print(f"There are {n_train} images for training")
```
